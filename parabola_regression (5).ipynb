## Least Squares Regression

### Polynomial Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Following the form $y = ax^2 + bx + c$
#### Set $a=3$, $b=2$, $c=10$

```python
N = 1000
x = np.arange(-15,15,30/N)*1.0
y = 3*(x**2) + 2*x + 10
plt.plot(x,y)
```

#### Graph of $y = 3x^2 + 2x + 10$

### Now plot random points around that line:

```python
np.random.seed(3)
mean = 0
std_deviation = 20
noise = np.random.normal(mean, std_deviation, N)
```

```python
y_noise = y+noise
plt.scatter(x,y_noise, s = 3)
```

## Construct Vandermonde Matrix
##### In the example for simple linear regression when we tried to find $\beta_0$ and $\beta_1$ in $y = \beta_0 + \beta_1 x$, we could use form $AX = b$ like this:

$\begin{bmatrix}
x_1 & 1 \\
x_2 & 1 \\
\vdots & \vdots \\
x_n & 1 \\
\end{bmatrix}\begin{bmatrix}
\beta_1 \\
\beta_0 \\
\end{bmatrix}
=
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n \\
\end{bmatrix}$


##### For multiple data points $(x_1, y_1), (x_2, y_2), \dots, (x_n, y_n)$ 
##### Similarly, once again we can use $AX = b$, except this time A is in the form

$\begin{bmatrix}
x_1^2 & x_1 & 1 \\
x_2^2 & x_2 & 1 \\
\vdots & \vdots & \vdots \\
x_n^2 & x_n & 1 \\
\end{bmatrix}$   

##### This matrix is called a <u>Vandermonde matrix<u>

##### Now, we have $AX = b$ as
$\begin{bmatrix} x_1^2 & x_1 & 1 \\ x_2^2 & x_2 & 1 \\ \vdots & \vdots & \vdots \\ x_n^2 & x_n & 1 \\ \end{bmatrix}\begin{bmatrix}a \\ b \\ c \\ \end{bmatrix} = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \\ \end{bmatrix}$

##### Where each row of b is calculated through $y_i = ax_i^2 + bx_i + c$


```python
b = y_noise
A = np.zeros((1000,3))
A[:,0] = x**2; A[:,1] = x; A[:,2] = 1
```

#### Use same formula $(A^TA)^{-1}A^Tb$  to find coefficient vector $X$

```python
AT = np.transpose(A)
ATA_inverse = np.linalg.inv(np.dot(AT,A))
ATb = np.dot(AT,b)
X = np.dot(ATA_inverse, ATb)
```

```python
y_calc = X[0]*x**2+X[1]*x+X[2]
plt.scatter(x,y_noise, s = 3)
plt.plot(x,y,color = "black", linewidth = 6)
plt.plot(x,y_calc, color = "red", linewidth = 2)
plt.show()
print(X)
```

#### The red line shows our calculated line, and the black line shows the actual function. 
#### Cool! Now we can make lines of best fit for parabolas too!
